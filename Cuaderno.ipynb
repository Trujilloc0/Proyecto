## Ejercicio 30
Escribe una función que reciba una lista de números y devuelva una tupla:
- cantidad de positivos,
- cantidad de negativos,
- cantidad de ceros.

````python
def tupla_counts(números_list):
  count_pos = 0
  count_neg = 0
  count_cero = 0
  for num in números_list:
    if num > 0:
      count_pos += 1
    elif num < 0:
      count_neg += 1
    else:
      count_cero += 1
  return (count_pos, count_neg, count_cero)

lista_input_size=int(input("Ingrese el tamaño de lista: "))
números=[]
for i in range(lista_input_size):
  prueba=int(input("Ingrese números:"))
  números.append(prueba)

result = tupla_counts(números)
print(result)
````

## Ejercicio 31
Escribe una función que determine si una lista está ordenada de manera no decreciente.

````python
def orden(input_list):

  if len(input_list) <= 1:
    return True
  for i in range(len(input_list) - 1):
    if input_list[i] > input_list[i+1]:
      return False
  return True

lista_input_size = int(input("Ingrese el tamaño de lista: "))
números = []
for i in range(lista_input_size):
  prueba = int(input("Ingrese números:"))
  números.append(prueba)

if orden(números):
  print("La lista está ordenada de manera creciente.")
else:
  print("La lista está ordenada de manera decreciente.")
````

## Ejercicio 37
Supón que llegan clientes en distintos tiempos.  
Diseña una función que reciba una lista de nombres y devuelva el orden exacto de atención.

````python
def orden_atencion(clientes):
    atendidos = []
    while len(clientes) > 0:
        cliente_actual = clientes.pop(0)
        atendidos.append(cliente_actual)
    return atendidos

nombres_llegada = ["Ana", "Juan", "Luis"]
print("Orden en el que llegaron:", nombres_llegada.copy())

resultado = orden_atencion(nombres_llegada)
print("Orden exacto de atención:", resultado)
````

# 9. Árboles binarios

Como el grupo ya llevó un curso de algoritmos, aquí trabajaremos una versión introductoria de árboles binarios usando clases sencillas.

````python
class Nodo:
    def __init__(self, valor, izquierdo=None, derecho=None):
        self.valor = valor
        self.izquierdo = izquierdo
        self.derecho = derecho
````

Construiremos este árbol:

```text
        8
       / \
      3   10
     / \    \
    1   6    14
```

````python
raiz = Nodo(
    8,
    Nodo(3, Nodo(1), Nodo(6)),
    Nodo(10, None, Nodo(14))
)

print("raíz:", raiz.valor)
print("hijo izquierdo:", raiz.izquierdo.valor)
print("hijo derecho:", raiz.derecho.valor)
````

## Recorridos clásicos
- **preorden**: raíz, izquierdo, derecho
- **inorden**: izquierdo, raíz, derecho
- **postorden**: izquierdo, derecho, raíz

## Ejercicio 38
Completa la función de recorrido en preorden.

````python
def preorden(nodo):
    if nodo is None:
        return
    print(nodo.valor)
    preorden(nodo.izquierdo)
    preorden(nodo.derecho)

preorden(raiz)
````

## Ejercicio 39
Completa la función de recorrido en inorden.

````python
def inorden(nodo):
    if nodo is None:
        return
    inorden(nodo.izquierdo)
    print(nodo.valor)
    inorden(nodo.derecho)

inorden(raiz)
````

## Ejercicio 40
Completa la función de recorrido en postorden.
